# Azure Portal 이용

이전 단계에서 설명한 [Python](https://learn.microsoft.com/ko-kr/azure/search/samples-python), [Rest API](https://learn.microsoft.com/ko-kr/rest/api/searchservice/?view=rest-searchservice-2024-07-01)과 같은 방법 이외에도 Azure Portal에서 UI로 이와 같은 작업들을 수행 할 수 있습니다.

### Search Service 생성

- [Azure Portal](https://learn.microsoft.com/ko-kr/azure/search/samples-python)에 [AI Search](https://portal.azure.com/#view/Microsoft_Azure_ProjectOxford/CognitiveServicesHub/~/CognitiveSearch) 서비스로 이동합니다.</p>
![Azure Portal AI Search](images/portal_ai_search_01.png)

- Search Service를 생성하기 위해 `Create` 버튼을 클릭합니다.</p>
![Azure Portal AI Search create service button](images/portal_ai_search_service_create_button.png)

- Basic 탭에서 Subscription, Resource Group, Service name, Location, Pricing tier를 입력 후 `Review+create`를 클릭합니다.</br>
(이 예제는 테스트용이며, 실제 서비스를 위해 생성한다면, Pricing tier, Scale, Networking 등 서비스를 위해 알맞는 설정을 하는 것이 좋습니다.)</p>
![Azure Portal AI Search create service](images/portal_ai_search_create_service.png)</br>
서비스 이름은 endpoint url로 사용되기에 다음의 조건을 만족하여야 합니다.
  - search.windows.net 네임스페이스 내에서 고유함
  - 2~60자 길이
  - 소문자, 숫자 또는 대시로 구성됨(-)
  - 처음 두 문자 또는 마지막 한 문자에 대시 사용 금지
  - 어디서든 연속 대시를 사용할 수 없음

- 그 후 이상이 없다면 `Create` 버튼을 클릭하여 서비스를 생성합니다.</p>
![Azure Portal AI Search create service](images/portal_ai_search_create_service_create.png)


[검색 서비스 만들기](https://learn.microsoft.com/ko-kr/azure/search/search-create-service-portal)에서 더 많은 내용들을 확인 할 수 있습니다.

### 인덱스(Index) 생성

- 서비스가 생성되면 `Go to resource` 버튼을 클릭하여 서비스로 이동하거나 AI Serach로 이동하여 조금 전 생성한 Serch Service를 선택합니다.
- 좌측의 `Overview` 선택 후 상단의 `Add index` 버튼을 클릭하여 인덱스를 추가 할 수 있습니다.
`Add index` 버튼을 클릭하여 나오는 두가지 메뉴에서 `Add index`는 UI를 통하여 필드(Field), Autocomplete 등 여러 설정들을 쉽게 입력 할 수 있으며, `Add index(JSON)`은 Json을 통하여 설정을 한 번에 입력 할 수 있습니다.</p>
![AI Search Add Index](images/portal_ai_search_add_index_button.png)</br>
- 여기에서는 `Add index(JSON)`을 클릭하여 index 생성 화면으로 넘어간 후 아래의 json을 복사하여 붙여 넣은 후 `Save` 버튼을 클릭하여 index를 생성합니다.<p>
![AI Search Add Index Json](images/portal_ai_search_add_index_json.png)</br>
- 그 후 생성 된 index를 클릭하여 세부 화면으로 이동후 Field등 index 생성 시 설정한 내용들을 확인 할 수 있으며, Web UI 환경에서 원하는 내용을 쉽게 진행 할 수 있습니다.</p>
![AI Search Add Index Json](images/portal_ai_search_index.png)</p>
![AI Search Add Index Detail](images/portal_ai_search_add_index_detail.png)</br>
